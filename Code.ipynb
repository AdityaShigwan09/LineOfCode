```python
#Write a python program in 3D to rotate the point (1, 0, 0) through XZ− plane in clockwise direction (rotation through Y− axis by an angle of 90◦).

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

point = np.array([1, 0, 0])

theta = np.radians(90)
rotation_matrix = np.array([[np.cos(theta), 0, np.sin(theta)],
                            [0, 1, 0],
                            [-np.sin(theta), 0, np.cos(theta)]])

rotated_point = np.dot(rotation_matrix, point)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(point[0], point[1], point[2], color='blue', label='Original Point (1, 0, 0)')

ax.scatter(rotated_point[0], rotated_point[1], rotated_point[2], color='red', label='Rotated Point')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])

# Add a legend
ax.legend()
plt.title('Rotation of (1, 0, 0) through XZ-plane (Clockwise by 90 degrees)')
plt.show()
```

```python
# Write a python program to rotate the segment by 180◦ having end points (1, 0) and (2,−1).

import numpy as np
import matplotlib.pyplot as plt

p1 = np.array([1, 0])
p2 = np.array([2, -1])

midpoint = (p1 + p2) / 2

theta = np.radians(180)
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                            [np.sin(theta), np.cos(theta)]])

p1_rotated = np.dot(rotation_matrix, p1 - midpoint) + midpoint
p2_rotated = np.dot(rotation_matrix, p2 - midpoint) + midpoint

plt.plot([p1[0], p2[0]], [p1[1], p2[1]], label='Original Segment')

plt.plot([p1_rotated[0], p2_rotated[0]], [p1_rotated[1], p2_rotated[1]], label='Rotated Segment')

plt.scatter(midpoint[0], midpoint[1], color='red', label='Midpoint')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Rotation of Segment by 180 degrees')
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Write a python program to draw a polygon with 8 sides having radius 5 centered at origin and find its area and perimeter.

import numpy as np
import matplotlib.pyplot as plt

num_sides = 8

radius = 5

angles = np.linspace(0, 2 * np.pi, num_sides + 1)[:-1]

x_coords = radius * np.cos(angles)
y_coords = radius * np.sin(angles)

plt.figure(figsize=(6, 6))
plt.plot(x_coords, y_coords, 'bo-')
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Regular Polygon with 8 Sides (Octagon)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

perimeter = num_sides * 2 * radius

area = 0.5 * num_sides * radius ** 2 * np.sin(2 * np.pi / num_sides)

print("Perimeter of the octagon:",perimeter)
print("Area of the octagon:",area)
print("\n")

plt.show()
```

```python
# Write a python program to find the area and perimeter of the ΔXY Z, where X(1, 2), Y (2,−2),Z(−1, 2).

import numpy as np

X = np.array([1, 2])
Y = np.array([2, -2])
Z = np.array([-1, 2])

def distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2) ** 2))


side_XY = distance(X, Y)
side_YZ = distance(Y, Z)
side_ZX = distance(Z, X)


perimeter = side_XY + side_YZ + side_ZX


s = perimeter / 2


area = np.sqrt(s * (s - side_XY) * (s - side_YZ) * (s - side_ZX))

print("Perimeter of triangle XYZ:",perimeter)
print("Area of triangle XYZ:", {area})
```

```python
'''
(i)Write a program to solve the following LPP:
Min Z = 3.5x + 2y
subject to x + y ≥ 5
x ≥ 4
y ≤ 2
x ≥ 0, y ≥ 0.
'''
import numpy as np
from scipy.optimize import linprog

c = [3.5, 2]


A = [[-1, -1],  
     [-1, 0],   
     [0, 1]]    

b = [-5, -4, 2]


x_bounds = (0, None)  
y_bounds = (0, None)  

res = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds], method='highs')

x = res.x[0]
y = res.x[1]
min_value = res.fun

print("Optimal Solution:")
print(f"x = {x}")
print(f"y = {y}")
print(f"Minimum value of Z = {min_value}")
```

```python
'''
(ii) Write a python program to solve LPP and find optimal solution if exists.
Max Z = 3x + 5y + 4z
subject to 2x + 3y ≤ 8
2y + 5z ≤ 10
3x + 2y + 4z ≤ 15
x ≥ 0, y ≥ 0, z ≥ 0.
'''

import numpy as np
from scipy.optimize import linprog


c = [-3, -5, -4]  

A = [[2, 3, 0],   
     [0, 2, 5],   
     [3, 2, 4]]   


b = [8, 10, 15]

x_bounds = (0, None)  
y_bounds = (0, None)  
z_bounds = (0, None)  
res = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds, z_bounds], method='highs')

if res.success:
    
    x = res.x[0]
    y = res.x[1]
    z = res.x[2]
    max_value = -res.fun  

    print("Optimal Solution:")
    print(f"x = {x}")
    print(f"y = {y}")
    print(f"z = {z}")
    print(f"Maximum value of Z = {max_value}")
else:
    print("No feasible solution exists.")
```

```python
'''
(ii) Write a python program to find the combined transformation on the line segment between the
points A[4, 1] & B[−3, 0] for the following sequence of Transformations:
(I) Rotation about origin through an angle π/4 .
(II) Uniform scaling by 7.3 units.
(III) Scaling in X− coordinate by 3 units.
(IV) Shearing in X direction by 1/2 units.
'''

import numpy as np

# Define the endpoints of the line segment A[4, 1] and B[-3, 0]
A = np.array([4, 1])
B = np.array([-3, 0])

# (I) Rotation about the origin through an angle π/4
theta = np.pi / 4
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                            [np.sin(theta), np.cos(theta)]])

A_rotated = np.dot(rotation_matrix, A)
B_rotated = np.dot(rotation_matrix, B)

# (II) Uniform scaling by 7.3 units
scaling_factor = 7.3
A_scaled = A_rotated * scaling_factor
B_scaled = B_rotated * scaling_factor

# (III) Scaling in X-coordinate by 3 units
scaling_factor_x = 3
A_scaled_x = np.array([A_scaled[0] * scaling_factor_x, A_scaled[1]])
B_scaled_x = np.array([B_scaled[0] * scaling_factor_x, B_scaled[1]])

# (IV) Shearing in X direction by 1/2 units
shearing_factor_x = 1 / 2
A_sheared = np.array([A_scaled_x[0] + shearing_factor_x * A_scaled_x[1], A_scaled_x[1]])
B_sheared = np.array([B_scaled_x[0] + shearing_factor_x * B_scaled_x[1], B_scaled_x[1]])

print("(I) Rotation about the origin through an angle π/4:")
print("A_rotated:", A_rotated)
print("B_rotated:", B_rotated)
print("\n(II) Uniform scaling by 7.3 units:")
print("A_scaled:", A_scaled)
print("B_scaled:", B_scaled)
print("\n(III) Scaling in X-coordinate by 3 units:")
print("A_scaled_x:", A_scaled_x)
print("B_scaled_x:", B_scaled_x)
print("\n(IV) Shearing in X direction by 1/2 units:")
print("A_sheared:", A_sheared)
print("B_sheared:", B_sheared)
```
